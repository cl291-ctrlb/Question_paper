```python
#Q1 (e) Discuss the advantages of using regression plots for predictive analysis.
#importing libraries
import seaborn as sns # statistical visualisation like regression plots
import matplotlib.pyplot as plt # plotting grahs and customizing plots
import pandas as pd # for tabular data as data frames

#creating sample data as dictionary
data={ 
    'Experience':[1,2,3,4],                   #independent variable 
    'Salary':[30000,45000,50000,55000]            #dependent variable 
}
df = pd.DataFrame(data) # convert dict to pandas dataframe

#create a regression plot
sns.regplot(x="Experience", y="Salary",data=df) # plots scatter plot of data and fits linear regression line t
plt.title("experince vs salary regression plot")
plt.xlabel("year of experience")
plt.ylabel("Salary")
plt.show()
```
